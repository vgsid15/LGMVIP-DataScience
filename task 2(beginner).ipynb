# applying minmax scalar

```python
import numpy as np
import sklearn
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
df1 = scaler.fit_transform(np.array(df1).reshape(-1,1))
```

```python
df1
```

# training and testing the dataset

```python
train_size = int(len(df1)*0.6)
test_size = len(df1) - train_size 
train_data,test_data = df1[0:train_size,:],df1[train_size:len(df1),:1]
```

```python
train_data,train_size
```

```python
test_data,test_size
```

```python

# conversion of training data and test data into a matrix
def create_dataset(dataset,time_step):
    dataX,dataY=[],[]
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step),0]
        dataX.append(a)
        dataY.append(dataset[i+time_step,0])
    return np.array(dataX) , np.array(dataY)

time_step = 100
X_train , y_train = create_dataset(train_data,time_step)
X_test , y_test = create_dataset(test_data,time_step)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# building a model

```python
import keras
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(100,1)))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mse', optimizer='adam')
```

```python
model.summary()
```

TRAINING THE DATA AND PREDICTING THE OUTPUT

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=40, batch_size=64)
```

```python
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
trans = model.predict(X_test[-1::]) 
output = scaler.inverse_transform(trans)
output
```

```python
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
```

```python
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_train, train_predict))
```

```python
math.sqrt(mean_squared_error(y_test, test_predict))
```
